```python
# Libraries import
import os
from urllib.parse import urlparse, parse_qs
from youtube_transcript_api import YouTubeTranscriptApi
from google.colab import userdata
import google.generativeai as genai
```

```python
# set API Key
GOOGLE_API_KEY = userdata.get('Gemini_API_Key')
genai.configure(api_key = GOOGLE_API_KEY)
```

```python
# install required package
# !pip install -q youtube-transcript-api
# prompt Definition
prompt =""" You are a summarizer. You will be taking the  test
   and summerizing the entire text and providing the important summary in points within 550 words.
   Please provide the summary of the  given text here : """
```

```python
# Genrate Gemini Content
def generate_gemini_content(transcript_text, prompt):
  model = genai.GenerativeModel("gemini-2.5-pro")
  response = model.generate_content(prompt + transcript_text)
  return response.text
```

```python
# colab input
user_input = input("Enter the text: ")
summary = generate_gemini_content(user_input, prompt)
print("\n## Summary")
print(summary)
```

```python
from fpdf import FPDF

# Assuming 'summary' variable contains the text to be converted to PDF
# You might need to install fpdf: !pip install fpdf

pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size = 12)

# Add the summary content to the PDF
# PyPDF2 is mainly for reading and manipulating existing PDFs, not creating from scratch easily with text flow.
# Using fpdf for easier text to PDF creation.

# Split text into lines to fit within the page width
# A simple approach, might need more sophisticated word wrapping for long text
lines = summary.split('\n')
for line in lines:
    pdf.cell(200, 10, txt = line, ln = True, align = 'L')


# Save the PDF
pdf_output_path = "summary.pdf"
pdf.output(pdf_output_path)

print(f"Summary saved to {pdf_output_path}")
```
