##**Python Logic Building Problems with Physics Context**

###**Problem 1: Calculate Distance Using Speed and Time**
Problem Statement: Write a program to calculate the distance traveled by an object, given
its speed and the time it travels.
Formula:
Distance = Speed × Time
Input: Speed (in meters per second) and time (in seconds).
Output: Distance traveled (in meters).
Example:
Input: Speed = 10, Time = 5
Output: Distance = 50 meters

```python
# Problem: Calculate Distance Using Speed and Time
# Physics Context: According to the basic formula of motion in physics,
# Distance = Speed × Time. This formula assumes constant speed and straight-line motion.

# Step 1: Take speed input from the user
# We use float to allow decimal values (e.g., 10.5 m/s)
speed = float(input("Enter speed in meters per second (m/s): "))

# Step 2: Take time input from the user
# Again, float is used to allow partial seconds (e.g., 2.5 seconds)
time = float(input("Enter time in seconds (s): "))

# Step 3: Apply the physics formula to calculate distance
# This is a direct multiplication of speed and time
distance = speed * time

# Step 4: Display the result with units
# We use f-string for clean formatting and clarity
print(f"Distance traveled = {distance} meters")
```

#**Problem 2: Calculate Force Using Newton’s Second Law**
Problem Statement: Write a program to calculate the force applied on an object using
Newton’s second law of motion.
Formula:
Force = Mass × Acceleration
Input: Mass (in kilograms) and acceleration (in meters per second squared).
Output: Force (in Newtons).
Example:
Input: Mass = 5, Acceleration = 3
Output: Force = 15 N


```python
# Problem 2: Calculate Force Using Newton’s Second Law
# Physics Context: Newton's Second Law states that Force is the product of an object's mass and its acceleration.
# Formula: Force = Mass × Acceleration

# Step 1: Take mass input from the user
# We use float to allow decimal values (e.g., 5.5 kg)
mass = float(input("Enter mass in kilograms (kg): "))

# Step 2: Take acceleration input from the user
# Acceleration is in meters per second squared (m/s²)
acceleration = float(input("Enter acceleration in meters per second squared (m/s²): "))

# Step 3: Apply Newton's formula to calculate force
# This is a direct multiplication of mass and acceleration
force = mass * acceleration

# Step 4: Display the result with units
# We use f-string for clean formatting and clarity
print(f"Force applied = {force} Newtons (N)")
```

#**Problem 3: Gravitational Potential Energy**
Problem Statement: Write a program to calculate the gravitational potential energy of an
object.
Formula:
Potential Energy = m × g × h
where g = 9.8 m/s2 (acceleration due to gravity), m is the mass, and h is the height.
Input: Mass (in kilograms) and height (in meters).

Output: Gravitational potential energy (in joules).
Example:
Input: Mass = 2, Height = 10
Output: Potential Energy = 196 J

```python
# Problem 3: Gravitational Potential Energy
# Physics Context: Gravitational potential energy is the energy stored due to an object's position above the ground.
# Formula: Potential Energy = m × g × h
# where:
# m = mass in kilograms (kg)
# g = acceleration due to gravity (9.8 m/s²)
# h = height in meters (m)

# Step 1: Take mass input from the user
# We use float to allow decimal values (e.g., 2.5 kg)
mass = float(input("Enter mass in kilograms (kg): "))

# Step 2: Take height input from the user
# Height is the vertical position above ground in meters
height = float(input("Enter height in meters (m): "))

# Step 3: Define gravitational acceleration constant
# This is a fixed value used in most physics problems on Earth
g = 9.8  # acceleration due to gravity in m/s²

# Step 4: Apply the formula to calculate potential energy
# Multiply mass, gravity, and height
potential_energy = mass * g * height

# Step 5: Display the result with units
# We use f-string for clean formatting
print(f"Gravitational Potential Energy = {potential_energy} Joules (J)")
```

#**Problem 4: Kinetic Energy of an Object**
Problem Statement: Write a program to calculate the kinetic energy of an object in motion.
Formula:
Kinetic Energy = 1/2 m v2
where m is the mass and v is the velocity.
Input: Mass (in kilograms) and velocity (in meters per second).
Output: Kinetic energy (in joules).
Example:
Input: Mass = 10, Velocity = 3
Output: Kinetic Energy = 45 J

```python
# Problem 4: Kinetic Energy of an Object
# Physics Context: Kinetic energy is the energy an object possesses due to its motion.
# Formula: Kinetic Energy = (1/2) × mass × velocity²

# Step 1: Take mass input from the user
# We use float to allow decimal values (e.g., 10.5 kg)
mass = float(input("Enter mass in kilograms (kg): "))

# Step 2: Take velocity input from the user
# Velocity is the speed of the object in a specific direction
velocity = float(input("Enter velocity in meters per second (m/s): "))

# Step 3: Apply the kinetic energy formula
# We square the velocity and multiply by half the mass
kinetic_energy = 0.5 * mass * velocity ** 2

# Step 4: Display the result with units
# We use f-string for clean formatting and clarity
print(f"Kinetic Energy = {kinetic_energy} Joules (J)")
```

#**Problem 5: Ohm’s Law – Voltage Calculation**
Problem Statement: Write a program that calculates the voltage across a resistor given
current and resistance using Ohm’s Law.
Formula:
V = I × R
where V is voltage, I is current, and R is resistance.
Input: Current (in amperes) and resistance (in ohms).
Output: Voltage (in volts).
Example:
Input: Current = 2, Resistance = 5
Output: Voltage = 10 V

```python
# Problem 5: Ohm’s Law – Voltage Calculation
# Physics Context: Ohm’s Law states that voltage (V) across a resistor is the product of current (I) and resistance (R).
# Formula: V = I × R

# Step 1: Take current input from the user
# Current is measured in amperes (A), and we use float to allow decimal values
current = float(input("Enter current in amperes (A): "))

# Step 2: Take resistance input from the user
# Resistance is measured in ohms (Ω), and float allows values like 5.5 ohms
resistance = float(input("Enter resistance in ohms (Ω): "))

# Step 3: Apply Ohm’s Law to calculate voltage
# This is a direct multiplication of current and resistance
voltage = current * resistance

# Step 4: Display the result with units
# We use f-string for clean formatting and clarity
print(f"Voltage across the resistor = {voltage} Volts (V)")
```

#**Problem 6: Projectile Motion – Maximum Height**
Problem Statement: Write a program to calculate the maximum height reached by a
projectile, given its initial velocity and launch angle (ignoring air resistance).
Formula:
h_max = (v2 × sin2(θ)) / (2g)
where v is the initial velocity, θ is the launch angle in degrees, and g = 9.8 m/s2.

Input: Initial velocity (in meters per second) and angle (in degrees).
Output: Maximum height (in meters).
Example:
Input: Velocity = 20, Angle = 45
Output: Height = 20.4 meters

```python
# Problem 6: Projectile Motion – Maximum Height
# Physics Context: When an object is launched at an angle, it follows a curved path called projectile motion.
# The highest point it reaches is called the maximum height.
# Formula: h_max = (v² × sin²(θ)) / (2 × g)
# where:
# v = initial velocity in meters per second (m/s)
# θ = launch angle in degrees
# g = gravitational acceleration (9.8 m/s²)

import math  # We need the math module to use sine and convert degrees to radians

# Step 1: Ask the user to enter the initial velocity
# This is the speed at which the object is launched
velocity = float(input("Please enter the initial velocity (in m/s): "))

# Step 2: Ask the user to enter the launch angle
# Angle should be in degrees, like 30°, 45°, or 60°
angle_degrees = float(input("Please enter the launch angle (in degrees): "))

# Step 3: Convert the angle from degrees to radians
# Python's math.sin() function works with radians, not degrees
angle_radians = math.radians(angle_degrees)

# Step 4: Define the gravitational acceleration constant
# This value is standard for Earth's surface
g = 9.8

# Step 5: Use the formula to calculate maximum height
# We square the velocity and sine of the angle, then divide by 2g
height = (velocity ** 2) * (math.sin(angle_radians) ** 2) / (2 * g)

# Step 6: Show the result to the user
# We round the result to two decimal places for clarity
print(f"The maximum height reached by the projectile is: {height:.2f} meters")
```

# **Problem 7: Harmonic Oscillator – Period of a Spring**
Problem Statement: Write a program to calculate the period of a spring-mass system
(simple harmonic oscillator).
Formula:
T = 2π √(m/k)
where T is the period, m is the mass, and k is the spring constant.
Input: Mass (in kilograms) and spring constant k (in N/m).
Output: Period T (in seconds).
Example:
Input: Mass = 1, Spring constant = 50
Output: Period = 0.89 s

```python
# Problem 7: Harmonic Oscillator – Period of a Spring
# Physics Context: In a simple harmonic oscillator, the period (T) is the time it takes to complete one full oscillation.
# Formula: T = 2π × √(m / k)
# where:
# m = mass in kilograms (kg)
# k = spring constant in Newtons per meter (N/m)
# π = mathematical constant (approximately 3.1416)

import math  # We import the math module to use π and square root functions

# Step 1: Prompt the user to enter the mass of the object
# Mass should be in kilograms (kg)
mass = float(input("Please enter the mass of the object (in kg): "))

# Step 2: Prompt the user to enter the spring constant
# Spring constant should be in Newtons per meter (N/m)
spring_constant = float(input("Please enter the spring constant (in N/m): "))

# Step 3: Apply the formula to calculate the period
# We use math.pi for π and math.sqrt() for square root
period = 2 * math.pi * math.sqrt(mass / spring_constant)

# Step 4: Display the result with appropriate units
# The result is rounded to two decimal places for clarity
print(f"The period of oscillation is: {period:.2f} seconds")
```

#**Problem 8: Free Fall – Time to Hit the Ground**
Problem Statement: Write a program to calculate the time it takes for an object to hit the
ground when dropped from a certain height (ignoring air resistance).
Formula:
t = √(2h/g)
where h is the height and g = 9.8 m/s2.
Input: Height (in meters).
Output: Time to hit the ground (in seconds).
Example:
Input: Height = 20
Output: Time = 2.02 seconds

```python
# Problem 8: Free Fall – Time to Hit the Ground
# Physics Context: When an object is dropped from a height, it accelerates downward due to gravity.
# The time it takes to reach the ground depends on the height and gravitational acceleration.
# Formula: t = √(2h / g)
# where:
# h = height in meters (m)
# g = acceleration due to gravity (9.8 m/s²)
# t = time in seconds (s)

import math  # We import the math module to use square root

# Step 1: Prompt the user to enter the height from which the object is dropped
# Height should be in meters (m)
height = float(input("Please enter the height from which the object is dropped (in meters): "))

# Step 2: Define the gravitational acceleration constant
# This value is standard for Earth's surface
g = 9.8

# Step 3: Apply the formula to calculate time
# We multiply height by 2, divide by g, then take the square root
time = math.sqrt((2 * height) / g)

# Step 4: Display the result with appropriate units
# The result is rounded to two decimal places for clarity
print(f"The time to hit the ground is: {time:.2f} seconds")
```

#**Problem 9: Work Done by a Force**
Problem Statement: Write a program to calculate the work done by a constant force acting
on an object.
Formula:
Work = F × d × cos(θ)

where F is the force, d is the displacement, and θ is the angle between the force and
displacement vectors.
Input: Force (in newtons), displacement (in meters), and angle (in degrees).
Output: Work done (in joules).
Example:
Input: Force = 100, Displacement = 10, Angle = 30
Output: Work = 866 J

```python
# Problem 9: Work Done by a Force
# Physics Context: Work is defined as the energy transferred when a force moves an object through a displacement.
# Formula: Work = F × d × cos(θ)
# where:
# F = force in newtons (N)
# d = displacement in meters (m)
# θ = angle between force and displacement vectors (in degrees)

import math  # We import the math module to use cosine and convert degrees to radians

# Step 1: Prompt the user to enter the magnitude of the force
# Force should be in newtons (N)
force = float(input("Please enter the force applied (in newtons): "))

# Step 2: Prompt the user to enter the displacement
# Displacement is the distance moved by the object in meters
displacement = float(input("Please enter the displacement (in meters): "))

# Step 3: Prompt the user to enter the angle between force and displacement
# Angle should be in degrees, like 0°, 30°, 90°, etc.
angle_degrees = float(input("Please enter the angle between force and displacement (in degrees): "))

# Step 4: Convert the angle from degrees to radians
# Python's math.cos() function expects the angle in radians
angle_radians = math.radians(angle_degrees)

# Step 5: Apply the formula to calculate work done
# Multiply force, displacement, and cosine of the angle
work_done = force * displacement * math.cos(angle_radians)

# Step 6: Display the result with appropriate units
# The result is rounded to the nearest whole number for clarity
print(f"The work done by the force is: {round(work_done)} Joules (J)")
```

#**Problem 10: Power Consumption of a Device**
Problem Statement: Write a program to calculate the power consumed by an electrical
device.
Formula:
P = W / t
where P is power, W is work (energy), and t is time.
Input: Work (in joules) and time (in seconds).
Output: Power (in watts).
Example:
Input: Work = 500, Time = 50
Output: Power = 10 W

```python
# Problem 10: Power Consumption of a Device
# Physics Context: Power is the rate at which energy is used or transferred.
# Formula: P = W / t
# where:
# W = work or energy in joules (J)
# t = time in seconds (s)
# P = power in watts (W)

# Step 1: Prompt the user to enter the amount of work or energy consumed
# Work should be entered in joules (J)
work = float(input("Please enter the amount of work done or energy used (in joules): "))

# Step 2: Prompt the user to enter the time duration
# Time should be entered in seconds (s)
time = float(input("Please enter the time duration (in seconds): "))

# Step 3: Apply the formula to calculate power
# Divide work by time to get power in watts
power = work / time

# Step 4: Display the result with appropriate units
# The result is rounded to two decimal places for clarity
print(f"The power consumed by the device is: {power:.2f} watts (W)")
```
